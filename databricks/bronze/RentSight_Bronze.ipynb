## ü•â BRONZE ‚Äî Ingest√£o do CSV ‚Üí Parquet (schema = raw)

A ideia √© **ler tudo como string** (mundo real) e salvar em Parquet para a pr√≥xima camada.


```python
from pyspark.sql import functions as F

raw_csv_path = '/Volumes/rentsightt/raw/airbnb_RJ.csv'
bronze_out_path = '/Volumes/rentsight/bronze/listings_bronze_rj'

print('RAW:', raw_csv_path)
print('BRONZE OUT:', bronze_out_path)
```

```python
df_raw = (
    spark.read
    .option('header', True)
    .option('inferSchema', False)
    .csv(raw_csv_path)
)

display(df_raw.limit(5))
print('Rows:', df_raw.count())
```

```python
# Salva como Parquet (Bronze)
(
    df_raw
    .write
    .mode('overwrite')
    .parquet(bronze_out_path)
)
print('‚úÖ Bronze gravado em:', bronze_out_path)
```
